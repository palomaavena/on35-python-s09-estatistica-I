```python
#1.	Calcular a média, mediana, mínimo, máximo e desvio padrão dos valores das vendas e quantidade vendida.

# 1º importando a biblioteca pandas
import pandas as pd


# função describe Pandas
df.describe().T
```

```python
#1.1	Calcular a média, mediana, mínimo, máximo e desvio padrão dos valores das vendas e quantidade vendida.

#Média
Média = print(f"A média de vendas é R$ {round(df['Valor da venda'].mean(),2)} e a média da quantidade vendida é {round(df['Quantidade vendida'].mean(),2)}")

#Mediana
Mediana = print(f"A mediana de vendas é R${round(df['Valor da venda'].median(),2)} e a mediana da quantidade vendida é {round(df['Quantidade vendida'].median(),2)}")

#Mínimo
Mínimo = print(f"O menor valor de vendas é R$ {round(df['Valor da venda'].min(),2)} e a menor quantidade vendida é {round(df['Quantidade vendida'].min(),2)}")

#Máximo
Máximo = print(f"O maior valor de vendas é R$ {round(df['Valor da venda'].max(),2)} e a maior quantidade vendida é {round(df['Quantidade vendida'].max(),2)}")

#Desvio Padrão
print(f"O desvio padrão de vendas é R$ {round(df['Valor da venda'].std(),2)} e o desvio padrão da quantidade vendida é {round(df['Quantidade vendida'].std(),2)}")
```

```python
#2.	Qual produto que mais vendeu? (Usar a moda para identificar).

#Moda
print(f"O produto que mais vendeu é {df['Produto'].mode()}")
```
